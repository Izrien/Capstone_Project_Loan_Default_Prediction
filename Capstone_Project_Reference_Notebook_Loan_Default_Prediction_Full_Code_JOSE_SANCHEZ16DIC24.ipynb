# **Loan Default Prediction**

## **Problem Definition** ###

Loan default prediction is a predictive modeling task in which a financial institution aims to estimate the likelihood that a borrower will fail to meet their scheduled loan obligations. More formally:

Given a set of historical loan applications and their associated outcomes (i.e., whether the borrower eventually defaulted or fully repaid),
A range of borrower attributes such as demographic information, credit history, financial stability indicators, and loan characteristics (e.g., interest rate, loan amount, loan term), as well as potentially external data like macroeconomic factors or industry trends.

- Objective:

**Develop a predictive model that, for a new loan applicant, outputs a probability or classification indicating the risk of default.**

- Key Considerations:

Accuracy vs. Practical Impact: The model should not only be statistically accurate but also provide insights that support effective credit risk management and decision-making.

Regulatory and Ethical Standards: The prediction should adhere to compliance standards (e.g., fair lending laws) and avoid unfair discrimination against protected groups.

Timeliness and Scalability: Predictions must be timely enough to support real-time or near real-time credit decisions and scale across large applicant volumes.

Interpretability and Actionability: The model’s outputs should be interpretable so that credit officers and risk managers can understand the reasons behind a given risk assessment and take appropriate actions, such as adjusting lending terms or initiating loss mitigation strategies.

In essence, the problem is about leveraging historical data and advanced analytics to preemptively identify potential defaulters, thereby mitigating financial risk and informing better lending decisions.

### **The Context:** ###

 **Loan default prediction is critically important for financial institutions for several key reasons:**

- Risk Mitigation and Loss Reduction:
Financial institutions earn revenue primarily by lending money and collecting interest. However, when borrowers default, the institution not only loses the principal amount but also incurs additional costs such as legal expenses and loan restructuring efforts. By accurately predicting which borrowers are likely to default, banks can proactively minimize their exposure to high-risk loans. This leads to a reduction in non-performing assets (NPAs) and helps maintain a healthier loan portfolio.

- Capital Efficiency and Regulatory Compliance:
Regulatory frameworks require banks to maintain sufficient capital reserves based on the risk profile of their lending activities. Accurate default predictions enable more precise estimation of risk-weighted assets and the capital required. This means the institution can allocate its capital more efficiently, ensuring compliance with regulations while not over-allocating capital that could be used productively elsewhere.

- Pricing and Product Structuring:
When a financial institution understands a borrower’s likelihood of default, it can tailor loan terms and interest rates more effectively. High-risk borrowers might face slightly higher interest rates or additional collateral requirements, while low-risk borrowers can enjoy more favorable terms. Accurate predictions thus allow for more competitive and personalized product offerings, ultimately improving market positioning and profitability.

- Informed Credit Policy and Underwriting Decisions:
Default prediction models provide data-driven insights that can shape lending policies. By integrating predictive analytics into credit decision-making, underwriters can better identify red flags, validate the borrower’s creditworthiness, and refine lending criteria. Over time, this leads to more consistent decision-making and a lower default rate.

- Improved Customer Relationship and Retention:
Default prediction isn’t just about denying credit to risky borrowers; it’s also about identifying early warning signs so the institution can intervene. For instance, if a model forecasts potential financial distress, the bank might offer a loan modification, financial counseling, or restructuring options before the situation deteriorates. Such proactive engagement can build trust, enhance the customer’s experience, and improve long-term customer relationships.

- Enhanced Strategic Planning and Forecasting:
A better understanding of default risk contributes to more accurate long-term planning. Institutions can simulate different economic scenarios—like downturns or changes in interest rates—and gauge how their loan portfolios might perform under stress. This forward-looking capability aids in strategic decision-making, optimizing loan portfolios, and preparing contingencies.


### **The objective:** ###

 The main objective of the loan default prediction project should be to develop a reliable, data-driven model that accurately estimates the probability of each borrower defaulting on their loan. This objective can be broken down into the following actionable goals:

Deliver predictions efficiently and at scale, integrating seamlessly with existing workflows and systems, and allowing for quick responses to changes in economic conditions or applicant profiles.

In essence, the objective is not merely to predict defaults accurately but to create a practical tool that supports smarter, more proactive credit risk management, ultimately improving both financial performance and customer relationships.

## **Data Description:**
The Home Equity dataset (HMEQ) contains baseline and loan performance information for 5,960 recent home equity loans. The target (BAD) is a binary variable that indicates whether an applicant has ultimately defaulted or has been severely delinquent. This adverse outcome occurred in 1,189 cases (20 percent). 12 input variables were registered for each applicant.


* **BAD:** 1 = Client defaulted on loan, 0 = loan repaid

* **LOAN:** Amount of loan approved.

* **MORTDUE:** Amount due on the existing mortgage.

* **VALUE:** Current value of the property. 

* **REASON:** Reason for the loan request. (HomeImp = home improvement, DebtCon= debt consolidation which means taking out a new loan to pay off other liabilities and consumer debts) 

* **JOB:** The type of job that loan applicant has such as manager, self, etc.

* **YOJ:** Years at present job.

* **DEROG:** Number of major derogatory reports (which indicates a serious delinquency or late payments). 

* **DELINQ:** Number of delinquent credit lines (a line of credit becomes delinquent when a borrower does not make the minimum required payments 30 to 60 days past the day on which the payments were due). 

* **CLAGE:** Age of the oldest credit line in months. 

* **NINQ:** Number of recent credit inquiries. 

* **CLNO:** Number of existing credit lines.

* **DEBTINC:** Debt-to-income ratio (all your monthly debt payments divided by your gross monthly income. This number is one way lenders measure your ability to manage the monthly payments to repay the money you plan to borrow.

## Treating Outliers

To handle outliers effectively, we first need to detect them. Here’s the step-by-step approach we can take:

- 1. Identify Outliers

Numerical Features: Use boxplots, Z-scores, or the Interquartile Range (IQR) method to detect outliers.
Visual Inspection: Histograms and scatterplots can also highlight extreme values.
Outlier Analysis: Identify specific rows with extreme values for further investigation.

- 2. Decide on Treatment

Remove Outliers: If outliers represent data entry errors or anomalies unlikely to repeat, consider removing them.
Cap/Transform Values: Cap extreme values at a threshold or apply transformations (e.g., log or square root) to reduce their influence.
Keep Outliers: Retain outliers if they reflect real-world phenomena and are relevant for model predictions.

- 3. Reassess Data Distribution

After treatment, replot the distributions to confirm the changes.
Let’s start by identifying outliers in the numerical features of your dataset. The IQR method will be used and the visualization of the outliers for the key variables.

```python
# Columns setup
numerical_columns = ['LOAN', 'MORTDUE', 'VALUE', 'YOJ', 'DEROG', 'DELINQ', 'CLAGE', 'NINQ', 'CLNO', 'DEBTINC']

# Detect and visualize outliers using the IQR method
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)  # First quartile (25th percentile)
    Q3 = data[column].quantile(0.75)  # Third quartile (75th percentile)
    IQR = Q3 - Q1  # Interquartile range
    lower_bound = Q1 - 1.5 * IQR  # Lower bound for outliers
    upper_bound = Q3 + 1.5 * IQR  # Upper bound for outliers
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

# Visualize outliers for key numerical features
outlier_results = {}

for column in numerical_columns:
    outliers, lower_bound, upper_bound = detect_outliers_iqr(hmeq_data, column)
    outlier_results[column] = {"Lower Bound": lower_bound, "Upper Bound": upper_bound, "Outlier Count": len(outliers)}
    
    # Boxplot to visualize outliers
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=hmeq_data[column], color='skyblue')
    plt.title(f'Boxplot of {column} (with potential outliers)')
    plt.xlabel(column)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

outlier_df = pd.DataFrame(outlier_results).T
# Display the outlier detection summary
print("Outlier Detection Summary")
print(outlier_df)
```

#### **Observations on Outliers:**  ####

**Key Findings**

- **LOAN (Loan Amount):**

Outliers are present above $41,600, with 256 observations flagged.
These high values might represent larger loans, potentially valid but less frequent.

**MORTDUE (Mortgage Due):**

Outliers are above $159,306, with 234 observations flagged.
These could indicate substantial mortgages and may still reflect real-world scenarios.

**VALUE (Property Value):**

Outliers are above $200,447, with 320 observations flagged.
Large property values could be realistic but skew the distribution.

**YOJ (Years at Present Job):**

Outliers are above 28 years or below 0 (invalid), with 91 flagged.
Negative or zero values for YOJ may be data entry errors.

**DEROG (Number of Major Derogatory Reports):**

Many outliers (725) due to the wide range up to 10.
While extreme values, they may accurately represent financial hardship.

#### **Insights** ####

- **Extreme Values:**

Outliers for LOAN, MORTDUE, and VALUE could be capped at a reasonable maximum (e.g., the 95th percentile) if they distort model training.

- **Potential Errors:**

Negative or zero values in YOJ are likely invalid and require removal or imputation.
Distribution Skew:

Variables like DEROG may benefit from log or square root transformation to reduce skewness.

```python
# Cap outliers at the 95th percentile for each numerical column
capped_data = hmeq_data.copy()

for column in numerical_columns:
    upper_cap = hmeq_data[column].quantile(0.95)
    capped_data[column] = hmeq_data[column].clip(upper=upper_cap)

# Visualize the capped distributions using boxplots
for column in numerical_columns:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=capped_data[column], color='lightgreen')
    plt.title(f'Boxplot of {column} (After Capping Outliers at 95th Percentile)')
    plt.xlabel(column)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

capped_data.describe()
```

**Boxplot Observations:**

- LOAN and MORTDUE: The capped values now create a more compressed distribution, making models more robust to extreme cases.

- VALUE and DEBTINC: Skewness has been reduced, with outliers brought closer to typical ranges.
Variable Range Adjustments:

For instance, LOAN now has a maximum value of $40,000 instead of $89,900, and MORTDUE has been capped at $151,999.5.

**Optimize Threshold:**

Calculate a custom metric or weighted cost based on the confusion matrix for each threshold.
Select the threshold that minimizes the total cost or maximizes business utility.
Evaluate Performance:

Analyze the confusion matrix and classification report at the optimized threshold.

```python
# Define costs for False Positives (FP) and False Negatives (FN)
cost_fp = 1  # Example: Cost of a false positive (e.g., unnecessary loan rejection)
cost_fn = 5  # Example: Cost of a false negative (e.g., loan loss due to default)

# Evaluate thresholds based on business costs
costs = []
for threshold in thresholds:
    # Predict based on the threshold
    y_pred_threshold = (y_pred_proba >= threshold).astype(int)
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred_threshold).ravel()
    
    # Calculate total cost
    total_cost = (cost_fp * fp) + (cost_fn * fn)
    costs.append(total_cost)

# Find the threshold with the minimum cost
optimal_threshold_idx = np.argmin(costs)
optimal_threshold = thresholds[optimal_threshold_idx]
min_cost = costs[optimal_threshold_idx]

# Plot total cost vs. thresholds
plt.figure(figsize=(10, 6))
plt.plot(thresholds, costs, label="Total Cost", color='purple')
plt.axvline(optimal_threshold, color='red', linestyle='--', label=f"Optimal Threshold = {optimal_threshold:.2f}")
plt.xlabel("Threshold")
plt.ylabel("Total Cost")
plt.title("Total Cost vs. Threshold")
plt.legend()
plt.grid(True)
plt.show()

# Evaluate model performance at the optimal threshold
y_pred_optimal_threshold = (y_pred_proba >= optimal_threshold).astype(int)
conf_matrix_optimal = confusion_matrix(y_test, y_pred_optimal_threshold)
classification_report_optimal = classification_report(y_test, y_pred_optimal_threshold)

optimal_threshold, min_cost, conf_matrix_optimal, classification_report_optimal
```

### **Decision Tree - Hyperparameter Tuning**

* Hyperparameter tuning is tricky in the sense that **there is no direct way to calculate how a change in the hyperparameter value will reduce the loss of your model**, so we usually resort to experimentation. We'll use Grid search to perform hyperparameter tuning.
* **Grid search is a tuning technique that attempts to compute the optimum values of hyperparameters.** 
* **It is an exhaustive search** that is performed on the specific parameter values of a model.
* The parameters of the estimator/model used to apply these methods are **optimized by cross-validated grid-search** over a parameter grid.

**Criterion {“gini”, “entropy”}**

The function to measure the quality of a split. Supported criteria are “gini” for the Gini impurity and “entropy” for the information gain.

**max_depth** 

The maximum depth of the tree. If None, then nodes are expanded until all leaves are pure or until all leaves contain less than min_samples_split samples.

**min_samples_leaf**

The minimum number of samples is required to be at a leaf node. A split point at any depth will only be considered if it leaves at least min_samples_leaf training samples in each of the left and right branches. This may have the effect of smoothing the model, especially in regression.
